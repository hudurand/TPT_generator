```python
import sys
import os
sys.path.append(os.path.abspath('..'))

import pandas as pd
import numpy as np
from pathlib import Path
from TPT_generator_python import DataBucket, TPTFetcher, TPTGenerator
from TPT_generator_python.constants import RATINGS
```

```python
DATE = "2020-12-31"
CLIENT = "Dynasty"
ISIN = "LU1280365476"
SYMADJ = -0.48
SOURCE_DIR = Path("../data")
OUTPUT_DIR = './production/example'
ALL = slice(None)
```

```python
generator = TPTGenerator(DATE,
                         CLIENT,
                         OUTPUT_DIR,
                         SOURCE_DIR,
                         sym_adj=SYMADJ,
                         shareclass_isin=ISIN)
generator
```

```python
ratings = generator.data_bucket.get_instruments(['rating_moodys','rating_sp','rating_fitch'])
ratings = ratings.loc[~(ratings.isnull().all(axis=1))]
ratings.replace({"u": "",
                 "p": "",
                 " \*+": "",
                 " \*-": "",
                 "e": ""}, regex=True, inplace=True)
ratings.replace("WD", None, regex=True, inplace=True)
ratings
```

```python
def func(x, y):
    if y is not None: 
        return RATINGS[x.name][y]
    
def sort_and_select_func(list_val):
    list_val.sort()
    return list_val[-2]

CQS = ratings.apply(lambda x: x.apply(lambda y: func(x,y)))
CQS["CQS"] = CQS.apply(lambda x: x.dropna().tolist(), axis=1)
CQS["CQS"].values.sort()
CQS
```
